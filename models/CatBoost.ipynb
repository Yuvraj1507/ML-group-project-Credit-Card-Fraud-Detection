```python
import pandas as pd

file_path = r'D:\python\Credit_card_fraud_detection\creditcard.csv'
data = pd.read_csv(file_path)
print(data.head())
x = data.drop(['Class'], axis=1)
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from catboost import CatBoostClassifier
#create CatBoost Model
clf = CatBoostClassifier()
#fit to our data
clf.fit(x_train, y_train)
#generate class probabilities
y_pred = clf.predict(x_test, prediction_type='RawFormulaVal')
```

```python
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
print (roc_auc)
```

```python
plt.plot(100*false_positive_rate, 100*true_positive_rate, label= 'catBoost (area = {:.3f})'.format(roc_auc), linewidth=2)
plt.xlabel('False positives [%]')
plt.ylabel('True positives [%]')
plt.xlim([0,30])
plt.ylim([60,100])
plt.grid(True)
ax = plt.gca()
ax.set_aspect('equal')
plt.title('CatBoost Model Performance')
plt.legend(loc='best')
plt.show()
```
